```python
def compare_norm_stand (model_name, wave_num, splits):
    model = ModelByModelName[model_name]
    wi = Config.wave_info(wave_num)
    wd = Config.wave_data(wave_num)
    scores_NWS = []
    lda_scores_NWS = []
    scores_SS = []
    lda_scores_SS = []
    num_features = range(2, wd.n_features, 1)

    steps = len (splits)
    step = 0
    p = ProgressBar(steps)

    for split_num in splits:
        p.animate (step)
        step += 1

        # Read splits from files
        split = Split()
        split.load_wave_train_test_RS (wave_num, split_num)

        mrmr_weights = read_mrmr(wi.filename_base, split_num)

        split_NWS = Split().copy (split)
        split_NWS.norm_weigh_sort (mrmr_weights)

        split_SS = Split().copy (split)
        split_SS.stand_sort (mrmr_weights)

        score_NWS, lda_score_NWS = Score_Model(model, num_features, split_NWS)
        scores_NWS.append(score_NWS)
        lda_scores_NWS.append(lda_score_NWS)

        score_SS, lda_score_SS = Score_Model(model, num_features, split_SS)
        scores_SS.append(score_SS)
        lda_scores_SS.append(lda_score_SS)

    p.animate (step)

    avg_score_NWS = Average_Score(scores_NWS)
    lda_avg_score_NWS = Average_Score(lda_scores_NWS)
    avg_score_SS = Average_Score(scores_SS)
    lda_avg_score_SS = Average_Score(lda_scores_SS)

    Graph ('{} mRMR (W{},S{}-{})-Norm&Weigh'.format(model_name,wave_num,splits[0],splits[-1]),
        num_features, avg_score_NWS, lda_avg_score_NWS)
    Graph ('{} mRMR (W{},S{}-{})-Stand'.format(model_name,wave_num,splits[0],splits[-1]),
        num_features, avg_score_SS, lda_avg_score_SS)
```

```python
def score_num_feat (model_name, wave_num, splits):

    filename = 'train_feat_{}_mRMR_W{}_S{}-{}_graphdat.pickle'.format(model_name,wave_num,splits[0],splits[-1])
    if os.path.isfile(filename):
        with open(filename, 'rb') as f:
            rd = pickle.load(f)
            num_features, avg_score, lda_avg_score = rd[0],rd[1],rd[2]
            print ('read results from',filename)
    else:
        model = ModelByModelName[model_name]
        wi = Config.wave_info(wave_num)
        wd = Config.wave_data(wave_num)
        scores = []
        lda_scores = []
        num_features = range(2, wd.n_features, 1)

        steps = len (splits)
        step = 0
        p = ProgressBar(steps)

        for split_num in splits:
            p.animate (step)
            step += 1

            # Make splits based on random-seed
            split = Split()
            split.load_wave_train_test_RS (wave_num, split_num)

            # Read mRMR from file
            mrmr_weights = read_mrmr(wi.filename_base, split_num)

            split.norm_weigh_sort (mrmr_weights)

            score, lda_score = Score_Model(model, num_features, split)
            scores.append (score)
            lda_scores.append(lda_score)

        p.animate (step)

        avg_score = Average_Score(scores)
        lda_avg_score = Average_Score(lda_scores)
        with open(filename, 'wb') as f:
            pickle.dump([num_features, avg_score, lda_avg_score], f, 3)
        print ('saved results in',filename)

    Graph ('Train Feat. {}-mRMR (W{},S{}-{})-Norm&Weigh'.format(
        model_name,wave_num,splits[0],splits[-1]), num_features, avg_score, lda_avg_score)
```
