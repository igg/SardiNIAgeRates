# Scoring of Different Models with Wave 1 Data

This notebook contains most of the initial model comparison that would produce some of the top performing models. Several different classfication and regression algorithms were used in combination with different dimensionality reduction methods. Almost all of the tests followed the format:

1. Data is read from csv file and converted to python
2. Data is cleaned
3. Data is binned with the optimal binning strategy
4. Different train and test samples are saved

Then:

1. Feature Selection (and weighting)
2. Apply LDA to selected features if necessary
3. Scoring (R^2) of models with increasing number of ranked features
4. Graphing results

NOTE: Some of the tests follow an older version of the Functions.ipynb file, which is why some experiments are set up differently from later ones.


### Dimensionality Reduction Algorithms:

For information on algorithm parameters, see FUNCTIONS Notebook

PCA (Principal Component Analysis) orthgonally transforms features into linearly uncorrelated principal components
- http://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html

LDA (Linear Discriminant Analysis) uses discretized data to find linear correlations that separate classes, has n-1 discriminants
- http://scikit-learn.org/stable/modules/generated/sklearn.lda.LDA.html

Pearson Correlation Coefficient (R) is a weighting/scoring algorithm based off of correlation (adopted from WND-CHRM code)
- https://github.com/wnd-charm/wnd-charm

Fisher Correlation Coefficient is a weighting/scoring algorithm that uses inter-class and intra-class differences (WND-CHRM)
- https://github.com/wnd-charm/wnd-charm

mRMR (minimum Redundancy Maximum Relevance) is a weighting/scoring algorithm that uses categorical data
- http://penglab.janelia.org/proj/mRMR/
