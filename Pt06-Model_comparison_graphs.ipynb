## Gender Graphs

These experiments dealt with comparing the top models as they were tested with either all male particpants, all female, or a mixed group. Then we tested for equal sample size (meaning that the number of males used in the male experiment was equivalent to the numer of females in the female experiment and the number of participants in the mixed experiment). We also tested with the maximum numnber of participants available to the male, female, and mixed experiments. Results show that there was no benefit from separation of gender unless there were similar sample sizes.
